## Research Question

Are NYPD officers of certain races more likely to receive civilian complaints
than expected based on their representation within the police force?

Using NYPD personnel data and civilian complaint records, we will compare the proportion of complaints received by officers of each race to their proportion within the police officer population. This is a statistical inference task that will assess whether observed complaint distributions differ significantly from expected distributions proportionally. We will use simple regression and statistical tests like a chi-square test or regression complaint rate modeling to evaluate disproportionality.

### CCRB Complaints Dataset

The CCRB (Civilian Complaint Review Board) dataset contains records of civilian complaints filed against NYPD officers. Each row represents one closed complaint case and includes information about the officer involved, the type of allegation, the date of the incident, and the final disposition of the investigation. The dataset contains 33,358 complaint cases and 27 variables.

The key variables for this project are officer race, allegation type, disposition and incident date. Officer race is a categorical variable identifying the race of the officer involved in the complaint (White, Black, Hispanic, Asian, Native American). Allegation type describes the nature of the complaint, such as abuse of authority, use of force, or offensive language. Incident date is recorded as a calendar date and indicates when the complaint occurred. Disposition describes the outcome of the CCRB investigation and typically falls into categories such as substantiated, unsubstantiated, or exonerated. These metrics allow us to quantify complaint patterns across officer racial groups.

There are several important limitations to this dataset. Complaint data reflects reported incidents, not confirmed misconduct. Some misconduct may go unreported due to many reasons, some of which include lack of trust in reporting systems, or lack of awareness of the complaint process. Reporting rates may also vary across neighborhoods and communities, which introduces potential reporting bias. Additionally, some fields contain missing values, particularly in demographic related columns. Because of these concerns, complaint counts should not be interpreted as direct measures of officer behavior but rather as reported interactions that triggered formal complaints.

### NYC Population Demographics (ACS)

The NYC Population Demographics dataset contains demographic breakdowns for New York City's population at both the citywide and borough levels. This dataset contains approximately 8.5 million residents with demographic estimates across 6 regions and approximately 90 categories for each region.

The important variables are the race proportions: White, Black, Asian, Hispanic, American Indian, and Pacific Islander. These are provided as both counts and percentages for each borough and for NYC as a whole. The race categories are categorical variables that represent the racial composition of NYC's population.

The purpose of this dataset is to establish expected proportions for arrest analysis.The problems with this dataset include that all proportions and estimates are not exact and are given with margins of error, as they come from survey estimates rather than exact counts. Additionally, the dataset represents a snapshot from the 2012-2016 ACS 5-year estimates, which may not reflect current population demographics if used for more recent analyses.

```python
raw = pd.read_csv('data/00-raw/NYC_Population_Demographics.csv', header=None)

regions = ['Bronx', 'Brooklyn', 'Manhattan', 'Queens', 'Staten Island', 'New York City']
metrics = ['Estimate', 'Margin of Error', 'Percent', 'Percent MoE']
col_names = ['Subject', 'Subcategory']
for r in regions:
    for m in metrics:
        col_names.append(f'{r}_{m}')
df_acs = raw.iloc[5:].copy()
df_acs.columns = col_names[:df_acs.shape[1]]
df_acs['Subject'] = df_acs['Subject'].ffill()

nyc_cols = [c for c in df_acs.columns if c.startswith('New York City_')]
def to_numeric_clean(s):
    v = str(s).strip().replace(',', '').replace('+/-', '').replace('%', '').replace('*****', '').replace('(X)', '')
    try:
        return pd.to_numeric(v) if v and v != 'nan' else np.nan
    except Exception:
        return np.nan
for c in nyc_cols:
    df_acs[c] = df_acs[c].apply(to_numeric_clean)
df_acs = df_acs[df_acs[nyc_cols].notna().any(axis=1)]

df_nyc = df_acs[['Subject', 'Subcategory'] + nyc_cols].copy()
df_nyc = df_nyc.rename(columns={c: c.replace('New York City_', 'nyc_').replace(' ', '_') for c in nyc_cols})


print("Shape (tidy NYC-only):", df_nyc.shape)
df_nyc.head(12)
```

```python
# Missingness check
missing = df_nyc.isna().sum()
print("Missing counts per column:\n", missing)
print("\nRows with any missing:", df_nyc.isna().any(axis=1).sum())
```

```python
# Outliers
print("nyc_Estimate range:", df_nyc['nyc_Estimate'].min(), "–", df_nyc['nyc_Estimate'].max())
print("nyc_Percent range:", df_nyc['nyc_Percent'].min(), "–", df_nyc['nyc_Percent'].max())

outlier = df_nyc[(df_nyc['nyc_Percent'].notna()) & ((df_nyc['nyc_Percent'] < 0) | (df_nyc['nyc_Percent'] > 100))]
print("Rows with percent outside [0,100]:", len(outlier))
outlier_est = df_nyc[(df_nyc['nyc_Estimate'].notna()) & (df_nyc['nyc_Estimate'] < 0)]
print("Rows with negative estimate:", len(outlier_est))
```

```python
# Clean
numeric_cols = ['nyc_Estimate', 'nyc_Margin_of_Error', 'nyc_Percent', 'nyc_Percent_MoE']
df_clean = df_nyc[df_nyc[numeric_cols].notna().any(axis=1)].copy()

print("Cleaned shape:", df_clean.shape)
df_clean.head()
```
